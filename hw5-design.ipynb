```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.neighbors import NearestNeighbors
```

```python
df= pd.read_csv("homework_6.1.csv")
df.info()
```

```python
treated = df[df["X"] == 1].reset_index(drop=True)
untreated = df[df["X"] == 0].reset_index(drop=True)
effects = []
nn_control = NearestNeighbors(n_neighbors=1)
nn_control.fit(untreated[["Z"]])

dist, idx = nn_control.kneighbors(treated[["Z"]])

for i in range(len(treated)):
    control_y = untreated.iloc[idx[i][0]]["Y"]
    effect = treated.iloc[i]["Y"] - control_y
    effects.append(effect)

nn_treated = NearestNeighbors(n_neighbors=1)
nn_treated.fit(treated[["Z"]])

dist, idx = nn_treated.kneighbors(untreated[["Z"]])

for i in range(len(untreated)):
    treated_y = treated.iloc[idx[i][0]]["Y"]
    effect = treated_y - untreated.iloc[i]["Y"]
    effects.append(effect)

ATE = np.mean(effects)
print("ATE =", ATE)
```

```python
nn = NearestNeighbors(n_neighbors=1)
nn.fit(untreated[["Z"]])

dist, idx = nn.kneighbors(treated[["Z"]])

effects = []

for i in range(len(treated)):
    control = untreated.iloc[idx[i][0]]
    effect = treated.iloc[i]["Y"] - control["Y"]
    effects.append(effect)

ATT = np.mean(effects)

print("ATT =", ATT)
```
